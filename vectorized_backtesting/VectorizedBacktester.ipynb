# Vectorized Backtester

Vectorized backtesting involves efficiently evaluating multiple trading strategies using optimized mathematical operations. Hence, this approach acclerates calculations, leading to quicker and more comprehensive strategy evaluations. Through this method, we can promptly identify strategies with superior risk-adjusted returns, which facilitates informed decisions and can potentially enhance portfolio profitability. 


The analysis below will showcase the backtesting results of an array of financial strategies. These strategies will be executed and assessed through historical backtesting, providing valuable insights on their performance across various market conditions.

## SMA Crossover

```python
from SMACrossoverTest import SMACrossoverTest
```

```python
sma_crossover = SMACrossoverTest('EURUSD=X', start='2000-01-01', end='2023-08-01', tc=0.00007, Fast_SMA=75, Slow_SMA=150, granularity='1d')
```

```python
# Outputs returns in the following order: (Cumulative Startegy Returns, Cumulative Standard Returns)
sma_crossover.test_strategy()
```

Our SMA Crossover strategy with a 75-period Fast SMA and 150-period Slow SMA significantly outperforms the standard buy/hold with 1-day granularity, from Jan 2004 to Aug 2023. Our strategy more than doubles its initial value, whereas the standard returns fall in comparison to initial value.

```python
sma_crossover.plot_results()
```

In the graph, we can more clearly visualize the large difference between our strategy and a regular buy/hold, depicting SMA Crossover's 100%+ increase in returns in comparison to the standard buy/hold.

## Mean Reversion (Bollinger Bands)

```python
from BollingerBandsTest import BollingerBandsTest
```

```python
bollinger_bands = BollingerBandsTest('EURUSD=X', start='2000-01-01', end='2023-08-01', tc=0.00007, SMA=25, standard_deviations=2, granularity='1d')
```

```python
bollinger_bands.test_strategy()
```

Our Mean Reversion strategy with Bollinger Bands using a 25-period SMA and 2$\sigma$ wide bands, outperforms the standard buy/hold with 1-day granularity, from Jan 2004 to Aug 2023. We see strategy returns hovering around 1.45 times its starting value, and once again, the standard buy/hold has decreased in value.

```python
bollinger_bands.plot_results()
```

In the graph, it is evident that the Bollinger Bands strategy outperforms the buy/hold, however this is largely due to the enormous spike in 2009-2010. Otherwise, we see that the increases diminish in intensity and remain relatively consistent with the buy/hold. As a volatility based indicator, I would not use this strategy alone, but rather coupled with various other strategies. 

## Relative Strength Index (RSI)

```python
from RSITest import RSITest
```

```python
rsi = RSITest('EURUSD=X', start='2000-01-01', end='2023-08-01', tc=0.00007, window=14, buy_threshold=70, sell_threshold=30, granularity='1d')
```

```python
rsi.test_strategy()
```

```python
rsi.plot_results()
```

In the RSI graph, we can see that standard buy/hold and RSI returns shift their dominance over the years. Thus, this momentum indicator is not a reliable strategy to trade on, and should be combined with other strategies to increase performance and profitability.
