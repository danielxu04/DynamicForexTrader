# Vectorized Backtester

Vectorized backtesting involves efficiently evaluating multiple trading strategies using optimized mathematical operations. Hence, this approach acclerates calculations, leading to quicker and more comprehensive strategy evaluations. Through this method, we can promptly identify strategies with superior risk-adjusted returns, which facilitates informed decisions and can potentially enhance portfolio profitability. 


The analysis below will showcase the backtesting results of an array of financial strategies. These strategies will be executed and assessed through historical backtesting, providing valuable insights on their performance across various market conditions.

## SMA Crossover

```python
from SMACrossoverTest import SMACrossoverTest
```

```python
sma_crossover = SMACrossoverTest('EURUSD=X', start='2000-01-01', end='2023-08-01', tc=0.00007, Fast_SMA=75, Slow_SMA=150, granularity='1d')
```

```python
# Outputs returns in the following order: (Cumulative Startegy Returns, Cumulative Standard Returns)
sma_crossover.test_strategy()
```

Our SMA Crossover strategy with a 75-period Fast SMA and 150-period Slow SMA significantly outperforms the standard buy/hold with 1-day granularity, from Jan 2004 to Aug 2023. Our strategy more than doubles its initial value, whereas the standard returns fall in comparison to initial value.

```python
sma_crossover.plot_results()
```

In the graph, we can more clearly visualize the large difference between our strategy and a regular buy/hold, depicting SMA Crossover's 100%+ increase in returns in comparison to the standard buy/hold.
